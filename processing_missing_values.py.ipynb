```python
import pandas as pd
from sklearn.preprocessing import Imputer
import numpy as np

db = pd.read_csv("database_class.csv")
print(db.describe())
```

```python
db = db.replace({'?': np.nan})
print(db)
```

```python
values = {
    'cMinPrice': db['cMinPrice'].astype(float).mode(),
    'cMaxPrice': db['cMaxPrice'].astype(float).mode(),
    'cSumPrice': db['cSumPrice'].astype(float).mode(),
    'bMinPrice': db['bMinPrice'].astype(float).mode(),
    'bMaxPrice': db['bMaxPrice'].astype(float).mode(),
    'bSumPrice': db['bSumPrice'].astype(float).mode(),
    'bStep': db['bStep'].astype(float).mode()[0],
    'onlineStatus': db['onlineStatus'].mode()[0],
    'availability': db['availability'].mode()[0],
    'customerNo': db['customerNo'].mode()[0],
    'maxVal': db['maxVal'].mode()[0],
    'customerScore': db['customerScore'].mode()[0],
    'accountLifetime': db['accountLifetime'].mode()[0],
    'payments': db['payments'].astype(float).mode()[0],
    'age': db['age'].astype(float).mode()[0],
    'address': db['address'].mode()[0],
    'lastOrder': db['lastOrder'].mode()[0],
}
db_mode = db.fillna(values)
db_mode.to_csv('database_class_mode_.csv')
```
